```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("RadwaH/DreamBoothAgnes2").to("cuda")

prompt = "beautiful anime sks girl, wearing lab coat and glasses, holding a clipboard, standing inside a research facility, character portrait, 1 9 6 0 s, long hair, intricate, elegant, highly detailed, digital painting, artstation, concept art, smooth, sharp focus, illustration, art by wlop, charlie bowater and alexandra fomina"
image = pipe(prompt, num_inference_steps=250, guidance_scale=7.5).images[0]

image.save("agnes-db-school-1.png")
```
