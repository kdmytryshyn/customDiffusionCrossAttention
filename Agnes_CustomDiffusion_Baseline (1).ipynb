```python
import torch
from diffusers import DiffusionPipeline

pipe = DiffusionPipeline.from_pretrained("RadwaH/CustomDiffusionAgnes2", torch_dtype=torch.float16).to("cuda")
pipe.unet.load_attn_procs("RadwaH/CustomDiffusionAgnes2", weight_name="pytorch_custom_diffusion_weights.bin")
pipe.load_textual_inversion("RadwaH/CustomDiffusionAgnes2", weight_name="<new1>.bin")

prompt = "beautiful anime <new1> girl, wearing lab coat and glasses, holding a clipboard, standing inside a research facility, character portrait, 1 9 6 0 s, long hair, intricate, elegant, highly detailed, digital painting, artstation, concept art, smooth, sharp focus, illustration, art by wlop, charlie bowater and alexandra fomina"
image = pipe(prompt, num_inference_steps=100, guidance_scale=6.0, eta=1.0).images[0]

image.save("agnes-cd-school.png")
```
